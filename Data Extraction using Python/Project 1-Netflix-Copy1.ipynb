```python
# Create a separate dataframe with movie title and columns containing all actor names

cast_columns = movies_df['Cast'].str.split(', ', expand=True)

cast_columns.columns = [f'Actor_{i+1}' for i in range(cast_columns.shape[1])]

movies_with_actors_df = pd.concat([movies_df['Title'], cast_columns], axis=1)

movies_with_actors_df
```

```python
# Create a dataframe of movies whose names starting with 'B'

df_movies_starting_with_B = movies_df[movies_df['Title'].str.startswith('B')].reset_index(drop=True)

df_movies_starting_with_B
```

```python
# Which movies have duration more than 150 min

dates_df['Duration']= dates_df['Duration'].str.replace(' min', '').astype(int)

df_long_duration = dates_df[dates_df['Duration'] > 150].reset_index(drop=True)

df_long_duration
```
